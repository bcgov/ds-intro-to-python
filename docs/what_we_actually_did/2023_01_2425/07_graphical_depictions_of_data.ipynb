```python
from seaborn import load_dataset
```

```python
penguins = load_dataset('penguins')
penguins.sample(10)
```

```python
x = penguins['bill_length_mm']
y = penguins['bill_depth_mm']

plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Penguin Beaks')
plt.show()
```

```python
x = [1,2,3,4,5]
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
x = [1,2,3,4,5]

# plot data
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')

# add legend
plt.legend()

# show data
plt.show()
```

```python
adelie = penguins[penguins.species == 'Adelie']
chinstrap = penguins[penguins.species == 'Chinstrap']
gentoo = penguins[penguins.species == 'Gentoo']

# create our 3 plots
plt.scatter(adelie.bill_length_mm, adelie.bill_depth_mm, label='Adelie')
plt.scatter(chinstrap.bill_length_mm, chinstrap.bill_depth_mm, label='Chinstrap')
plt.scatter(gentoo.bill_length_mm, gentoo.bill_depth_mm, label='Gentoo')

# create labels
plt.xlabel('Bill Length')
plt.ylabel('Bill Depth')
plt.title('Beaks')

# create legend
plt.legend()

# show 
plt.show()
```
